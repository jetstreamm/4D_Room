```python
mashed_potato = potato.combinatorial_polyhedron()
```

```python
mashed_potato_graph = mashed_potato.vertex_graph()
```

```python
mashed_potato_edges = list(mashed_potato_graph.edges())
```

```python
six_hundred_cell_edges = []
for a, b, _ in mashed_potato_edges: 
    new_tuple = (tuple(a), tuple(b))
    six_hundred_cell_edges.append(new_tuple)
```
